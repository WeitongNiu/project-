```python
import pandas as pd

#Read in data
US_Export = pd.read_csv("Export2002.csv", encoding = 'utf-8')
US_Import = pd.read_csv("Import2002.csv", encoding = 'utf-8')

#Clean the Export Data

Data = {'Year':[], 'Period':[], "Value" :[]}
Months = ["January",'February','March','April','May','June','July','August','September','October','November','December']
for i in range(24):
    for j in Months:
        Data["Year"].append(US_Export.iloc[i]["Year"]) 
        Data["Period"].append(j)
        if type(US_Import.iloc[i][j]) != float:
            Data["Value"].append(int(US_Export.iloc[i][j].replace(",","")))
        else: 
            Data["Value"].append(US_Export.iloc[i][j])
Cleaned_Export = pd.DataFrame(Data).dropna().reset_index()

#Clean the Import Data

Data = {'Year':[], 'Period':[], "Value" :[]}
Months = ["January",'February','March','April','May','June','July','August','September','October','November','December']
for i in range(24):
    for j in Months:
        Data["Year"].append(US_Import.iloc[i]["Year"]) 
        Data["Period"].append(j)
        if type(US_Import.iloc[i][j]) != float:
            Data["Value"].append(int(US_Import.iloc[i][j].replace(",","")))
        else: 
            Data["Value"].append(US_Import.iloc[i][j])
Cleaned_Import = pd.DataFrame(Data).dropna().reset_index()

Cleaned_Import["Index"] = Cleaned_Import.index
Cleaned_Export["Index"] = Cleaned_Export.index

#df["Import"] = Cleaned_Import["Value"]
#df["Export"] = Cleaned_Export["Value"]
```

```python
Cleaned_Export.rename(columns={"Value": "Export_usa_monthly"}, inplace=True)
Cleaned_Export = Cleaned_Export.reset_index(drop=True)

Cleaned_Export = Cleaned_Export.iloc[:-1]
Cleaned_Export 
```

```python
Cleaned_Export.to_csv('US_Export_cleaned.csv', index=False)
```

```python
Cleaned_Import.rename(columns={"Value": "Import_usa_monthly"}, inplace=True)
Cleaned_Import = Cleaned_Import.iloc[:-1]
Cleaned_Import
```

```python
Cleaned_Import.to_csv('US_Import_cleaned.csv', index=False)
```
